# Split into 40 frames each action and train

```python
import os, cv2
import json
from utils import kth, streamer, frameCount


config = json.load(open("config.json", "r"))
kthPath = config["paths"]["kth_dataset"]
kth40Path = config["paths"]["kth40"]
chunkWidth = 40

def dumpBuffer(path, buffer):
    for idx, img in enumerate(buffer):
        cv2.imwrite('{}.jpg'.format(os.path.join(path, str(idx))),img)
    print("dump at {}".format(path))
                    
for activity in kth.listActivities(kthPath):
    
    # Create folder if not exists  
    activityPath = os.path.join(kth40Path, activity + "/")
    if not os.path.isdir(activityPath):
        os.mkdir(activityPath)
        print("{} folder created".format(activityPath))
        
    # Extract images from vid
    activityFiles = kth.listFiles(kthPath, activity, abosultePath=True)
    
    clipNumber = 0 
    for _f in activityFiles:
        noOfFrames = frameCount(_f)
        residueFrames = noOfFrames % chunkWidth
        buffer = [frame for frame in streamer(_f)]    
        clipsCount = noOfFrames // chunkWidth
        for clip in range(clipsCount):
            clipPath = os.path.join(kth40Path, activity, "clip{}".format(clipNumber))
            clipNumber += 1
            if not os.path.isdir(clipPath + "/"):
                os.mkdir(clipPath + "/")
                print("{} folder created".format(clipPath))
            dumpBuffer(clipPath, buffer[clip*chunkWidth:(clip+1)*chunkWidth])
        del buffer
            
```
